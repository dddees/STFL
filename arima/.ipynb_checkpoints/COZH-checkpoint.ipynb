```python

import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
def autoRegr(df, end_train,lag,option):
    end_train=pd.to_datetime(end_train,utc=True)
    day=pd.to_timedelta(1,'day')
    train_data=df.loc[df['Time']<end_train,'Load']
    test_data=df.loc[(df['Time']>=end_train)&(df['Time']<end_train+day),'Load']
    PSE_pred=df.loc[(df['Time']>=end_train)&(df['Time']<end_train+day),'PSEforecast'].values
    xlabel=np.arange(24)
    
    ar_model=AutoReg(train_data,lags=lag).fit()
    pred=ar_model.predict(test_data.index[0],test_data.index[-1]).values
    test_data=test_data.values

    mae=mean_absolute_error(test_data,pred)
    rmse=np.sqrt(mean_squared_error(test_data,pred))
    r2=r2_score(test_data,pred)
    mape=np.mean(np.abs((test_data-pred)/test_data))*100 #percentage
    smape=np.mean(np.abs(pred-test_data)/(0.5*np.abs(pred-test_data)))
    #plt.plot(xlabel,pred)
    #plt.plot(xlabel,test_data)
    #plt.plot(xlabel,PSE_pred)
    #plt.legend(['Model','Real','PSEpred'])
    #plt.title(end_train.date())
    #plt.grid()
    if option==0:
        return np.array([mae,rmse,r2,mape,smape])
    else:
        return pred

```

```python
metrics=np.zeros([40,5])
for date in pd.date_range(start='2022-02-28',end='2023-02-28'):
    for i in np.arange(40):
        metrics[i]+=(autoRegr(df,date,lag=i,0))
```

```python

metrics=metrics/366

fig,ax=plt.subplots(2)
ax[0].plot(metrics[:,0])
ax[0].plot(metrics[:,1])
ax[0].legend(['mae','rmse'])
ax[0].set_ylabel('Error [MW]')
ax[1].plot(metrics[:,2])
ax[1].plot(metrics[:,3])
ax[1].plot(metrics[:,4])
ax[1].set_ylabel('Error')

ax[1].legend(['r2','mape','smape'])
ax[0].set_xticks(np.arange(0,40,2))
ax[1].set_xticks(np.arange(0,40,2))


```
